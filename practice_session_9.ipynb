```python
i = 1

while i <= 50:
    if i%2 == 0:
        print(i)
    i += 1




# The above code will print even numbers from 2 to 50, one per line.

# The above code will not print any numbers because the condition i%2 == 0 is false for all odd numbers.
# This code prints numbers from 1 to 50, one per line.
# It uses a while loop that continues until i is greater than 50.
# The variable i is initialized to 1 and incremented by 1 in each iteration.
# The print statement outputs the current value of i in each iteration.
```

```python
### Pyhon program to take input from user and return factorial of that number
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number: "))
result = factorial(num)
print(f"The factorial of {num} is {result}")
# The above code will take a number as input from the user and return its factorial.
```

```python
n = int(input("Enter a number: "))
res = 1
while n>0:
    res = res*n
    n -= 1
print(res)
# The above code will take a number as input from the user and return its factorial.
```

```python
# Take input from the user
n = int(input("Enter a number: "))

# Convert the number to a string to iterate over its digits
number_str = str(n)

# Create an empty dictionary to store digit frequencies
digit_frequency = {}

# Iterate through each digit in the number
for digit in number_str:
    if digit in digit_frequency:
        digit_frequency[digit] += 1
    else:
        digit_frequency[digit] = 1

# Print the dictionary with digit frequencies
print("Digit frequencies:", digit_frequency)
```

```python
# Take input from the user
n = int(input("Enter a number: "))

# Create an empty dictionary to store digit frequencies
digit_frequency = {}

# Process each digit using a while loop
while n > 0:
    digit = n % 10  # Extract the last digit
    digit = str(digit)  # Convert digit to string for dictionary key
    if digit in digit_frequency:
        digit_frequency[digit] += 1
    else:
        digit_frequency[digit] = 1
    n = n // 10  # Remove the last digit

# Print the dictionary with digit frequencies
print("Digit frequencies:", digit_frequency)
```
